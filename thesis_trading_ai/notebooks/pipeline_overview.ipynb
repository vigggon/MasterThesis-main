# Thesis Trading AI — Pipeline Overview

1. **data_download.py** — MT5, NAS100/US100 M5 (2019–2025) → `data/raw/nasdaq_m5.csv`
2. **volatility_validation.py** — ATR(14), Q75, Open vs Rest (Mann–Whitney) → `data/volatility_analysis/`
3. **session_filter.py** — 09:30–11:30 NY → `data/session_filtered/nasdaq_open_session.csv`
4. **feature_engineering.py** — Returns, ATR, RSI, EMA, VWAP, etc. → `data/features/open_features.csv`
5. **label_generator.py** — SL=1×ATR, TP=1.2×ATR, max 12 bars → `data/labeled/open_labeled.csv`
6. **dataset_builder.py** — Window=24, train (2019–2023) / backtest (2024) / forward (2025) → `data/processed/*.npz`
7. **train.py** — LSTM / Transformer on train only
8. **evaluate_ml.py**, **stability_analysis.py**, **backtester.py**, **forward_test.py** — Metrics and simulations
9. **visualize.py** — Equity curves, performance backtest vs forward
